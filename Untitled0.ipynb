# **Financial Data Analysis**

# Extract

```python
#Import required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/Financial Analytics data.csv")
df.head()
```

```python
#check NUll values

df.info()
```

# Modify Data

```python
# Drop the 'Unnamed: 4' column as it seems irrelevant
df_cleaned = df.drop(columns=['Unnamed: 4'])

# Drop rows where 'Mar Cap - Crore' or 'Sales Qtr - Crore' are missing
df_cleaned = df_cleaned.dropna(subset=['Mar Cap - Crore', 'Sales Qtr - Crore'])

# Check the cleaned dataset
df_cleaned.info(), df_cleaned.head()
```

# Key Trends

```python
# Identify the top 10 companies based on Market Capitalization
top_10_market_cap = df_cleaned.nlargest(10, 'Mar Cap - Crore')

# Identify the top 10 companies based on Quarterly Sales
top_10_sales = df_cleaned.nlargest(10, 'Sales Qtr - Crore')

# Visualization: Market Capitalization vs. Quarterly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=df_cleaned)
plt.title('Market Capitalization vs. Quarterly Sales')
plt.xlabel('Market Capitalization (in Crores)')
plt.ylabel('Quarterly Sales (in Crores)')
plt.grid(True)
plt.show()

# Display the top 10 companies by Market Cap and Sales
top_10_market_cap[['Name', 'Mar Cap - Crore']], top_10_sales[['Name', 'Sales Qtr - Crore']]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Mar Cap - Crore', y='Name', data=top_10_market_cap)
plt.title('Top 10 Companies by Market Capitalization')
plt.xlabel('Market Capitalization (in Crores)')
plt.ylabel('Company')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='Sales Qtr - Crore', y='Name', data=top_10_sales)
plt.title('Top 10 Companies by Quarterly Sales')
plt.xlabel('Quarterly Sales (in Crores)')
plt.ylabel('Company')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (in Crores)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Sales Qtr - Crore'], bins=30, kde=True)
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Quarterly Sales (in Crores)')
plt.ylabel('Frequency')
plt.show()
```

# Correlation Matrices

```python
# Calculate descriptive statistics for Market Capitalization and Quarterly Sales
descriptive_stats = df_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].describe()

# Calculate the correlation between Market Capitalization and Quarterly Sales
correlation = df_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()

descriptive_stats, correlation
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
